```python
from groq import Groq


from openai import OpenAI
import os
client = OpenAI(
    api_key=userdata.get('STIS_GROQ_API_KEY2'),
    base_url="https://api.groq.com/openai/v1",
)

def paraphrase(input):
  response = client.responses.create(
      input=f'''Parafrase input berikut dengan mempertahankan format tag yang diberikan. Tampilkan hanya hasil parafrase tanpa tambahan text atau penjelasan lainnya.
      input:
      {input}
      ''',
      model="openai/gpt-oss-20b",
  )
  return response.output_text
```

```python
from groq import Groq


from openai import OpenAI
import os
client = OpenAI(
    api_key=userdata.get('STIS_GROQ_API_KEY2'),
    base_url="https://api.groq.com/openai/v1",
)

def reasoning(entitas1, hubungan, entitas2, ent1, ent2, reason):
  response = client.responses.create(
      input=f'''Jelaskan mengapa secara semantik "{entitas1}" {hubungan} terhadap "{entitas2}", seperti "{ent1}" {hubungan} terhadap "{ent2}" dengan alasan {reason}. Tampilkan hanya penjelasan inti tanpa tambahan text atau penjelasan lainnya.
      ''',
      model="openai/gpt-oss-20b",
  )
  return response.output_text
```

```python
import pandas as pd

datasets = {
    "survei_dataset_balanced_paraphrase": survei_dataset_balanced_paraphrase,
    "survei_dataset_QA_balanced_paraphrase": survei_dataset_QA_balanced_paraphrase,
}

# Gabungkan semua value_counts menjadi satu tabel
counts_table = pd.concat(
    {name: df["relationship"].value_counts(dropna=False) for name, df in datasets.items()},
    axis=1
).fillna(0).astype(int)

# (opsional) urutkan baris label sesuai total terbanyak
counts_table = counts_table.loc[counts_table.sum(axis=1).sort_values(ascending=False).index]

# (opsional) tambah total per kolom
counts_table.loc["TOTAL"] = counts_table.sum(axis=0)

display(counts_table)          # Jupyter/Colab
# print(counts_table.to_string())  # kalau mau versi print juga
```
